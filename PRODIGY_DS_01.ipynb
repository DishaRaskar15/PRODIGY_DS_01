```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Load the dataset
file_path = "C:/Users/Admin/Downloads/demographic_data.csv"  # Update this path if needed
df = pd.read_csv(file_path)
```

```python
# Select top 10 counties by total population
top10 = df.sort_values("Total Population", ascending=False).head(10)
```

```python

# Set up bar chart data
bar_width = 0.35
x = range(len(top10)) 
```

```python

# Plotting
plt.figure(figsize=(12, 6))
plt.bar(x, top10['Male Population'], width=bar_width, label='Male', color='skyblue')
plt.bar([i + bar_width for i in x], top10['Female Population'], width=bar_width, label='Female', color='lightcoral')

# Labels and title
plt.xlabel('County')
plt.ylabel('Population')
plt.title('Top 10 Most Populous Counties by Gender')
plt.xticks([i + bar_width / 2 for i in x], top10['County'], rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
